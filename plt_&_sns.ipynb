```python
import pandas as pd
auto = pd.read_csv('/content/Automobile_data (1).csv')
print(auto.to_string())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales=[1000,1223,2345,678,543]
profit=[200,210,300,120,100]
year=[1999,2000,2001,2002,2003]
plt.subplot(2,2,1)
plt.plot(sales,year)
plt.subplot(2,2,2)
plt.plot(profit,year)
```

```python
plt.plot(sales,year)
plt.xlabel('sales',color = 'red',fontsize = 'xx-large')
plt.ylabel('year',color = 'Blue',fontsize = 'xx-large')
plt.title('Sales and Year Graph')
plt.show()
```
